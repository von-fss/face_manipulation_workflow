```python
from retinaface import RetinaFace
import cv2
import matplotlib.pyplot as plt

imgPath = r'.\tests\dataset\source.jpg'
resp = RetinaFace.detect_faces(imgPath)
```

```python
resp
```

```python
resp.keys()
```

```python
facial_area = resp['face_1']['facial_area']
```

```python
facial_area
```

```python
img = cv2.imread(imgPath)
cv2.rectangle(img, (facial_area[2], facial_area[3]), (facial_area[0], facial_area[1]), (255,255,255), 1)
plt.figure(figsize=(5,5))
plt.imshow(img[:,:,::-1])
plt.show()
```

```python
extract_face = RetinaFace.extract_faces(imgPath, align=True, expand_face_area=20)
```
